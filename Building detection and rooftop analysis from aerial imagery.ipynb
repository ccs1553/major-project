**STEP3 : DATA PREPROCESSING**

```python
# slicing the original images and ground truths into tiles of 384 * 384 pixels
# each image will be divided into 19 * 19 = 361 tiles

for filename in src_train_images:
    print(filename)
    master_img = cv.imread(os.path.join(src_train_folder, filename))
    master_img_gt = cv.imread(os.path.join(src_train_folder_gt, filename))

    for i in range(count):
        if i < count - 1:
            y = round(i * step)
        else:
            y = master_size - image_size # last tile of the column

        for j in range(count):
            if j < count - 1:
                x = round(j * step)
            else:
                x = master_size - image_size # last tile of the row


# Slice the main image based on (x,y)  first go ====>>> img = master_img[0:0+384, 0:0+384] = img = master_img[0:384,0:384]
            img = master_img[y:y+image_size, x:x+image_size]
            img_gt = master_img_gt[y:y+image_size, x:x+image_size]

# write the slice (tile) into disk
# filename[:-4]  to get rid of file extension in original filename (ex: austin1.tif ==>>>  austin1 )

            img_fname = '{}_{}_{}.{}'.format(filename[:-4], i, j, 'jpg')
            img_gt_fname = '{}_{}_{}.{}'.format(filename[:-4], i, j, 'png')
            cv.imwrite(os.path.join(train_folder, img_fname), img)
            cv.imwrite(os.path.join(train_folder_gt, img_gt_fname), img_gt)
```

**Model1 U-net**

```python
from keras.layers import *
from tensorflow.keras.optimizers import *

# non-sequential model

input_size = (img_height,img_width,3)

# convolution layers and pooling for downsampling
inputs = Input(input_size)
conv1 = Conv2D(64, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(inputs)
conv1 = Conv2D(64, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(conv1)
pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)

conv2 = Conv2D(128, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(pool1)
conv2 = Conv2D(128, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(conv2)
pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)

conv3 = Conv2D(256, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(pool2)
conv3 = Conv2D(256, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(conv3)
pool3 = MaxPooling2D(pool_size=(2, 2))(conv3)

conv4 = Conv2D(512, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(pool3)
conv4 = Conv2D(512, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(conv4)
drop4 = Dropout(0.5)(conv4)
pool4 = MaxPooling2D(pool_size=(2, 2))(drop4)

conv5 = Conv2D(1024, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(pool4)
conv5 = Conv2D(1024, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(conv5)
drop5 = Dropout(0.5)(conv5)


# deconvolution layers and concatenation for upsampling
up6 = Conv2D(512, 2, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(UpSampling2D(size = (2,2))(drop5))
merge6 = concatenate([drop4,up6], axis = 3)
conv6 = Conv2D(512, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(merge6)
conv6 = Conv2D(512, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(conv6)

up7 = Conv2D(256, 2, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(UpSampling2D(size = (2,2))(conv6))
merge7 = concatenate([conv3,up7], axis = 3)
conv7 = Conv2D(256, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(merge7)
conv7 = Conv2D(256, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(conv7)

up8 = Conv2D(128, 2, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(UpSampling2D(size = (2,2))(conv7))
merge8 = concatenate([conv2,up8], axis = 3)
conv8 = Conv2D(128, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(merge8)
conv8 = Conv2D(128, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(conv8)

up9 = Conv2D(64, 2, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(UpSampling2D(size = (2,2))(conv8))
merge9 = concatenate([conv1,up9], axis = 3)
conv9 = Conv2D(64, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(merge9)
conv9 = Conv2D(64, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(conv9)
conv9 = Conv2D(2, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(conv9)

#final output layer
conv10 = Conv2D(1, 1, activation = 'sigmoid')(conv9)

model1 = Model( inputs, conv10)

#compiling the model
model1.compile(optimizer = Adam(learning_rate = 1e-4), loss = 'binary_crossentropy', metrics = ['accuracy'])



#Printing the model summary to see the architecture and the number of parameters
model1.summary()
```

```python
#creating a function to plot the training and validation accuracy

def plot_model(history):
  accuracy = history.history['accuracy']
  val_accuracy = history.history['val_accuracy']
  loss = history.history['loss']
  val_loss = history.history['val_loss']
  epochs_range = range(10)

  plt.figure(figsize=(8, 5))
  plt.subplot(1, 2, 1)
  plt.plot(epochs_range, accuracy, label='Training Accuracy')
  plt.plot(epochs_range, val_accuracy, label='Validation Accuracy')
  plt.legend(loc='lower right')
  plt.title('Training and Validation Accuracy')

  plt.subplot(1, 2, 2)
  plt.plot(epochs_range, loss, label='Training Loss')
  plt.plot(epochs_range, val_loss, label='Validation Loss')
  plt.legend(loc='upper right')
  plt.title('Training and Validation Loss')
  plt.show()

plot_model(results1)
```

**FPN with VGG16**

```python
import tensorflow as tf
from tensorflow.keras import layers, models, applications

def build_fpn_vgg(input_shape=(384, 384, 3), num_classes=1):
    # Loading VGG16 backbone,
    vgg16 = applications.VGG16(weights='imagenet', include_top=False, input_shape=input_shape)

    # Getting the feature maps from different stages of VGG16
    c1 = vgg16.get_layer("block1_conv2").output  # 192x192
    c2 = vgg16.get_layer("block2_conv2").output  # 96x96
    c3 = vgg16.get_layer("block3_conv3").output  # 48x48
    c4 = vgg16.get_layer("block4_conv3").output  # 24x24
    c5 = vgg16.get_layer("block5_conv3").output  # 12x12

    # Apply a 1x1 convolution to reduce the channel depth and start the FPN process
    p5 = layers.Conv2D(256, (1, 1), padding="same")(c5)
    p5_upsampled = layers.Conv2DTranspose(256, (2, 2), strides=(2, 2), padding="same")(p5)  # 12x12 -> 24x24

    p4 = layers.Conv2D(256, (1, 1), padding="same")(c4)
    p4 = layers.Add()([p5_upsampled, p4])
    p4_upsampled = layers.Conv2DTranspose(256, (2, 2), strides=(2, 2), padding="same")(p4)  # 24x24 -> 48x48

    p3 = layers.Conv2D(256, (1, 1), padding="same")(c3)
    p3 = layers.Add()([p4_upsampled, p3])
    p3_upsampled = layers.Conv2DTranspose(256, (2, 2), strides=(2, 2), padding="same")(p3)  # 48x48 -> 96x96

    p2 = layers.Conv2D(256, (1, 1), padding="same")(c2)
    p2 = layers.Add()([p3_upsampled, p2])
    p2_upsampled = layers.Conv2DTranspose(256, (2, 2), strides=(2, 2), padding="same")(p2)  # 96x96 -> 192x192

    p1 = layers.Conv2D(256, (1, 1), padding="same")(c1)
    p1 = layers.Add()([p2_upsampled, p1])

    # Ensure the output shape is correct
    outputs = layers.Conv2D(128, (3, 3), padding="same")(p1)

    # Final convolution layer for segmentation output with desired output shape
    outputs = layers.Conv2D(num_classes, (1, 1), activation="sigmoid", padding="same")(outputs)

    # Defining the model
    model2 = models.Model(inputs=vgg16.input, outputs=outputs)

    return model2

# Instantiating the model
model2 = build_fpn_vgg()

# Compiling the model
model2.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

# Displays the model summary
model2.summary()
```

```python
def fcn_model(input_size=(384, 384, 3)):
    inputs = Input(input_size)

    # Encoder layer (Downsampling path)
    conv1 = Conv2D(64, (3, 3), padding='same')(inputs)
    conv1 = BatchNormalization()(conv1)
    conv1 = Activation('relu')(conv1)
    conv1 = Conv2D(64, (3, 3), padding='same')(conv1)
    conv1 = BatchNormalization()(conv1)
    conv1 = Activation('relu')(conv1)
    pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)

    conv2 = Conv2D(128, (3, 3), padding='same')(pool1)
    conv2 = BatchNormalization()(conv2)
    conv2 = Activation('relu')(conv2)
    conv2 = Conv2D(128, (3, 3), padding='same')(conv2)
    conv2 = BatchNormalization()(conv2)
    conv2 = Activation('relu')(conv2)
    pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)

    conv3 = Conv2D(256, (3, 3), padding='same')(pool2)
    conv3 = BatchNormalization()(conv3)
    conv3 = Activation('relu')(conv3)
    conv3 = Conv2D(256, (3, 3), padding='same')(conv3)
    conv3 = BatchNormalization()(conv3)
    conv3 = Activation('relu')(conv3)
    pool3 = MaxPooling2D(pool_size=(2, 2))(conv3)

    # Bottleneck layer
    conv4 = Conv2D(512, (3, 3), padding='same')(pool3)
    conv4 = BatchNormalization()(conv4)
    conv4 = Activation('relu')(conv4)
    conv4 = Conv2D(512, (3, 3), padding='same')(conv4)
    conv4 = BatchNormalization()(conv4)
    conv4 = Activation('relu')(conv4)

    # Decoder layer(Upsampling path)
    upconv3 = Conv2DTranspose(256, (3, 3), strides=(2, 2), padding='same')(conv4)
    upconv3 = concatenate([upconv3, conv3])
    conv5 = Conv2D(256, (3, 3), padding='same')(upconv3)
    conv5 = BatchNormalization()(conv5)
    conv5 = Activation('relu')(conv5)
    conv5 = Conv2D(256, (3, 3), padding='same')(conv5)
    conv5 = BatchNormalization()(conv5)
    conv5 = Activation('relu')(conv5)

    upconv2 = Conv2DTranspose(128, (3, 3), strides=(2, 2), padding='same')(conv5)
    upconv2 = concatenate([upconv2, conv2])
    conv6 = Conv2D(128, (3, 3), padding='same')(upconv2)
    conv6 = BatchNormalization()(conv6)
    conv6 = Activation('relu')(conv6)
    conv6 = Conv2D(128, (3, 3), padding='same')(conv6)
    conv6 = BatchNormalization()(conv6)
    conv6 = Activation('relu')(conv6)

    upconv1 = Conv2DTranspose(64, (3, 3), strides=(2, 2), padding='same')(conv6)
    upconv1 = concatenate([upconv1, conv1])
    conv7 = Conv2D(64, (3, 3), padding='same')(upconv1)
    conv7 = BatchNormalization()(conv7)
    conv7 = Activation('relu')(conv7)
    conv7 = Conv2D(64, (3, 3), padding='same')(conv7)
    conv7 = BatchNormalization()(conv7)
    conv7 = Activation('relu')(conv7)

    outputs = Conv2D(1, (1, 1), activation='sigmoid')(conv7)

    model3 = Model(inputs=[inputs], outputs=[outputs])

    return model3
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import ResNet50

def conv_block(input_tensor, filters, kernel_size=3, strides=1, padding='same'):
    x = layers.Conv2D(filters, kernel_size=kernel_size, strides=strides, padding=padding)(input_tensor)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    return x

def upsample_block(input_tensor, skip_tensor, filters):
    x = layers.Conv2DTranspose(filters, kernel_size=3, strides=2, padding='same')(input_tensor)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    x = layers.Concatenate()([x, skip_tensor])
    x = conv_block(x, filters)
    return x

def build_linknet(input_shape=(384, 384, 3), num_classes=1):
    # Encoder - ResNet50
    resnet = ResNet50(weights='imagenet', include_top=False, input_shape=input_shape)

    # Extracting relevant feature maps from ResNet50
    encoder_output = resnet.get_layer("conv4_block6_out").output
    skip1 = resnet.get_layer("conv1_relu").output  # 64 filters
    skip2 = resnet.get_layer("conv2_block3_out").output  # 256 filters
    skip3 = resnet.get_layer("conv3_block4_out").output  # 512 filters

    # Decoder
    x = upsample_block(encoder_output, skip3, 512)
    x = upsample_block(x, skip2, 256)
    x = upsample_block(x, skip1, 64)

    x = layers.Conv2DTranspose(32, kernel_size=3, strides=2, padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)

    # Final Convolution
    output = layers.Conv2D(num_classes, kernel_size=1, activation='sigmoid')(x)

    # Model
    model4 = models.Model(inputs=resnet.input, outputs=output)
    return model4

# Model Summary
model4 = build_linknet()
model4.summary()
```
